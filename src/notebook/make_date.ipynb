```python
import pandas as pd
from datetime import datetime
import random

import pandas as pd
from datetime import datetime, timedelta
import random

# 钢材仓库列表
STEEL_INVENTORY_NAME_LIST = [
    "山西宏达钢铁有限公司厂库",
    "山西华鑫源钢铁集团有限公司厂库",
    "安阳市新普钢铁有限公司厂库",
    "安钢集团信阳钢铁有限责任公司厂库",
    "四川德润钢铁集团航达钢铁有限责任公司厂库",
    "山西晋南钢铁集团有限公司厂库",
    "山西新泰钢铁有限公司厂库",
    "云南曲靖呈钢钢铁(集团)有限公司厂库",
    "六安钢铁控股集团有限公司仓库",
    "山西建龙实业有限公司厂库",
    "中拓物流(武汉建投汉阳库)",
]
# 煤炭仓库列表
COAL_INVENTORY_NAME_LIST = [
    "杨桥畔华茂二号库",
    "靖江盈利港",
    "京唐港",
    "直发港",
    "太仓港",
]

# 公司名称列表
COMPANY_NAME_LIST = [
    "湖北国贸能源化工有限公司",
    "湖北国贸金属矿产有限公司",
    "湖北国贸汽车有限公司",
    "湖北国际贸易集团有限公司",
    "湖北国贸农产品有限公司",
    "武汉鼎联丰国际贸易有限公司",
    "湖北国贸农产品有限公司武汉分公司",
    "湖北南方大集实业有限公司",
    "湖北南方大集实业有限公司东西湖分公司",
    "湖北南方大集实业有限公司慈惠分公司",
    "湖北南方大集实业有限公司江汉分公司",
    "湖北南方大集实业有限公司能源分公司",
    "湖北南方工贸有限公司",
    "湖北南方集团有限公司",
    "湖北国贸供应链管理有限公司",
    "湖北华中能源发展有限公司",
    "湖北国贸汽车有限公司红安分公司",
]
import pandas as pd
import random
from datetime import datetime


# 为每个仓库随机分配一个公司
coal_companies = {
    warehouse: random.choice(COMPANY_NAME_LIST)
    for warehouse in COAL_INVENTORY_NAME_LIST
}
steel_companies = {
    warehouse: random.choice(COMPANY_NAME_LIST)
    for warehouse in STEEL_INVENTORY_NAME_LIST
}

# 初始化数据列表
coal_inventory_data = []
steel_inventory_data = []
throughput_data = []

# 生成煤炭库存数据
for warehouse, company in coal_companies.items():
    inventory_weight = round(random.uniform(1000, 50000), 2)
    estimated_value = inventory_weight * 700
    prepaid_goods = round(random.uniform(0, estimated_value * 0.8))
    coal_inventory_data.append(
        {
            "公司": company,
            "仓库": warehouse,
            "库存重量": inventory_weight,
            "暂估货值": estimated_value,
            "预付货款": prepaid_goods,
        }
    )

# 生成钢材库存数据
for warehouse, company in steel_companies.items():
    inventory_quantity = round(random.uniform(100, 1000), 2)
    inventory_weight = inventory_quantity * 7.85
    stock_value = round(inventory_weight * 4000)
    steel_inventory_data.append(
        {
            "公司": company,
            "仓库": warehouse,
            "库存数量": inventory_quantity,
            "库存重量": inventory_weight,
            "存货金额余额": stock_value,
        }
    )

# 生成仓库吞吐量数据
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 1)
# 使用集合来记录已经生成吞吐量数据的公司-仓库-月份组合
generated_throughputs = set()

for single_date in (
    start_date + timedelta(days=i) for i in range((end_date - start_date).days)
):
    # 对于每个公司和仓库的组合，检查是否已经为当前月份生成了吞吐量数据
    for companies in [coal_companies, steel_companies]:  # 遍历煤炭和钢材仓库
        for warehouse, company in companies.items():
            month_key = (company, warehouse, single_date.month, single_date.year)
            if month_key not in generated_throughputs:
                throughput = round(random.uniform(100, 1000), 2)
                throughput_data.append(
                    {
                        "公司": company,
                        "仓库": warehouse,
                        "月度吞吐量": throughput,
                        "月份": single_date.strftime("%Y年%m月"),
                    }
                )
                generated_throughputs.add(month_key)  # 记录已生成的组合


# 构造DataFrame
coal_inventory_df = pd.DataFrame(coal_inventory_data)
steel_inventory_df = pd.DataFrame(steel_inventory_data)
throughput_df = pd.DataFrame(throughput_data)

# 保存到Excel文件
with pd.ExcelWriter("inventory_and_throughput_data.xlsx", engine="openpyxl") as writer:
    coal_inventory_df.to_excel(writer, sheet_name="煤炭库存表", index=False)
    steel_inventory_df.to_excel(writer, sheet_name="钢材库存表", index=False)
    throughput_df.to_excel(writer, sheet_name="仓库吞吐量记录表", index=False)

print("库存和吞吐量数据已成功保存到Excel文件中。")
```
